```python
plt.figure(figsize = (15,7))
df['Category_Name'].value_counts().plot(kind='bar')
plt.title('Number of Orders by Category')
plt.xlabel('Category')
plt.ylabel('Number of Orders')
plt.show()
```

```python
plt.figure(figsize = (8, 8))
pivot_table = pd.pivot_table(df, values='Sales', index='Category_Name', columns='Customer_Segment', aggfunc=np.sum)
sns.heatmap(pivot_table, cmap='Blues')
plt.title('Total Sales by Category and Customer Segment')
plt.show()
```

```python
df.groupby(['Type', 'Order_Status']).size().unstack().plot(kind='bar')
plt.title('Number of Orders by Customer Segment and Delivery Status')
plt.xlabel('Customer Segment')
plt.ylabel('Number of Orders')
plt.show()
```

```python
g = sns.catplot(
    data=df,
    x="Latitude", y="Market", row="Late_delivery_risk",
    kind="box", orient="h",
    sharex=False, margin_titles=True,
    height=1.5, aspect=4,
)
g.set(xlabel="Latitude", ylabel="")
g.set_titles(row_template="{row_name} class")
for ax in g.axes.flat:
    ax.xaxis.set_major_formatter('{x:.0f}')
```

```python
cols = ["Shipping_Mode", "Sales_per_customer", "Order_Item_Quantity"]

# Group the data by the Shipping Mode feature
groups = df.groupby("Shipping_Mode")

# Create the 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
for name, group in groups:
    ax.scatter(group[cols[0]], group[cols[1]], group[cols[2]], label=name)
ax.set_xlabel(cols[0])
ax.set_ylabel(cols[1])
ax.set_zlabel(cols[2])
plt.title("3D Scatter Plot for Shipping Mode")
plt.legend(loc="best")

# Display the plot
plt.show()
```
